```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, BatchNormalization, ReLU, MaxPooling1D
from tensorflow.keras.layers import GlobalAveragePooling1D, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# --- 1. ECG Beat Template ---
def make_beat_template(fs=300):
    t = np.linspace(0, 1, fs)
    p = np.exp(-((t-0.2)/0.025)**2)
    qrs = -np.exp(-((t-0.3)/0.01)**2)*2 + np.exp(-((t-0.305)/0.016)**2)*5
    t_wave = 0.5*np.exp(-((t-0.5)/0.04)**2)
    beat = p + qrs + t_wave
    return beat

# --- 2. Generate Signal Function ---
def generate_ecg_sample(duration=5, fs=300, arrhythmia=False):
    n_samples = duration * fs
    beat = make_beat_template(fs)
    mean_rr = 0.8
    sd_rr = 0.05
    signal = np.zeros(n_samples)
    t_idx = 0
    while t_idx + fs < n_samples:
        rr = np.random.normal(mean_rr, sd_rr)
        rr += np.random.uniform(-0.03, 0.03)
        if arrhythmia and np.random.rand() < 0.5:
            rr *= np.random.uniform(0.3, 0.7)
            beat_mod = beat * np.random.uniform(0.5, 1.3)
        else:
            beat_mod = beat * np.random.uniform(0.9, 1.1)
        rr_samples = int(rr * fs)
        end = min(n_samples, t_idx + fs)
        signal[t_idx:end] += beat_mod[:end-t_idx]
        t_idx += rr_samples
    wander = 0.1 * np.sin(2*np.pi*0.5 * np.arange(n_samples)/fs)
    noise = np.random.normal(0, 0.05, n_samples)
    return signal + wander + noise

# --- 3. Build Dataset ---
N = 5000  # total samples per class
duration, fs = 5, 300
X, y = [], []
for cls, arr_flag in enumerate([False, True]):
    for _ in range(N):
        X.append(generate_ecg_sample(duration, fs, arrhythmia=arr_flag))
        y.append(cls)
X = np.array(X)[..., np.newaxis]
y = to_categorical(np.array(y))

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# --- 4. Model Definition ---
model = Sequential([
    Conv1D(32, 5, padding='same', input_shape=(duration*fs,1)), BatchNormalization(), ReLU(), MaxPooling1D(2),
    Conv1D(64, 5, padding='same'), BatchNormalization(), ReLU(), MaxPooling1D(2),
    Conv1D(128, 5, padding='same'), BatchNormalization(), ReLU(), MaxPooling1D(2),
    Conv1D(256, 5, padding='same'), BatchNormalization(), ReLU(), MaxPooling1D(2),
    GlobalAveragePooling1D(),
    Dense(64, activation='relu'), Dropout(0.3),
    Dense(2, activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# --- 5. Training ---
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=20,
    batch_size=32,
    callbacks=[early_stop]
)

# --- 6. Evaluation ---
print("\nTest Results:")
loss, acc = model.evaluate(X_test, y_test)
print(f"Loss={loss:.4f}, Accuracy={acc:.4f}")

y_pred = np.argmax(model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['Normal','Arrhythmia']))
print("\nConfusion Matrix:")
print(confusion_matrix(y_true, y_pred))

# --- 7. Plot Sample Signals & Training Curve ---
times = np.arange(duration*fs)/fs
plt.figure(figsize=(12,4))
for i, cls in enumerate([0,1]):
    plt.subplot(1,2,i+1)
    sample = X_test[y_true==cls][0].squeeze()
    plt.plot(times, sample)
    plt.title(['Normal','Arrhythmia'][cls])
    plt.xlabel('Time (s)')
plt.tight_layout()
plt.show()

plt.figure()
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Save model
model.save('ecg_arrhythmia_cnn.h5')
print("Model saved: ecg_arrhythmia_cnn.h5")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_ecg_beat(t, beat_start, fs, normal=True):
    """
    Generate one ECG beat at position beat_start.
    If normal=True, generate normal beat,
    else generate arrhythmia beat (irregular timing and shape).
    """
    # Normal beat components
    p_wave = np.exp(-((t - (beat_start + 0.1))**2) / (2 * (0.025**2))) * 0.1
    qrs = np.exp(-((t - (beat_start + 0.2))**2) / (2 * (0.01**2))) * 1.0
    t_wave = np.exp(-((t - (beat_start + 0.35))**2) / (2 * (0.05**2))) * 0.3

    if normal:
        beat = p_wave - qrs + t_wave
    else:
        # Arrhythmia: jitter beat timing and amplitudes, irregular QRS
        jitter = np.random.uniform(-0.05, 0.05)
        p_wave = np.exp(-((t - (beat_start + 0.1 + jitter))**2) / (2 * (0.03**2))) * (0.1 + 0.05*np.random.rand())
        qrs = np.exp(-((t - (beat_start + 0.2 + jitter))**2) / (2 * (0.015**2))) * (1.0 + 0.5*np.random.rand())
        t_wave = np.exp(-((t - (beat_start + 0.35 + jitter))**2) / (2 * (0.07**2))) * (0.3 + 0.2*np.random.rand())
        beat = p_wave - qrs + t_wave
    return beat

def generate_synthetic_ecg_sample(duration=10, fs=300, heart_rate=60, normal=True, noise_level=0.02):
    """
    Generate one synthetic ECG signal sample.
    """
    t = np.linspace(0, duration, int(duration * fs))
    beats_per_sec = heart_rate / 60
    beat_period = 1 / beats_per_sec

    ecg = np.zeros_like(t)
    beat_times = np.arange(0, duration, beat_period)

    if not normal:
        # Add some irregularity in beat intervals for arrhythmia
        beat_times = beat_times + np.random.uniform(-0.1, 0.1, size=beat_times.shape)
        beat_times = np.clip(beat_times, 0, duration)
        beat_times = np.sort(beat_times)

    for bt in beat_times:
        ecg += generate_ecg_beat(t, bt, fs, normal=normal)

    # Add baseline wander + noise
    baseline = 0.05 * np.sin(2 * np.pi * 0.5 * t)
    noise = np.random.normal(0, noise_level, len(t))
    ecg += baseline + noise

    return t, ecg

def generate_dataset(n_samples=100, duration=10, fs=300):
    """
    Generate a dataset of synthetic ECG signals and labels.
    Half are normal, half are arrhythmia.
    Returns: X (n_samples, signal_length), y (labels: 0=normal,1=arrhythmia)
    """
    X = []
    y = []
    for i in range(n_samples):
        normal = (i < n_samples // 2)
        # Heart rate varies randomly around 60-100 bpm
        hr = np.random.randint(60, 101)
        _, ecg = generate_synthetic_ecg_sample(duration=duration, fs=fs, heart_rate=hr, normal=normal)
        X.append(ecg)
        y.append(0 if normal else 1)
    return np.array(X), np.array(y)

# Example usage
X, y = generate_dataset(n_samples=6, duration=5, fs=300)

# Plot some samples with labels
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.plot(X[i])
    plt.title(f'Sample {i+1} - {"Normal" if y[i]==0 else "Arrhythmia"}')
    plt.xlabel('Sample Index')
    plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_ecg_beat(t, beat_start, fs, condition='normal'):
    """
    Generate one ECG beat at beat_start with a given condition.
    condition: 'normal', 'arrhythmia', 'st_depression', 'st_elevation', 't_wave_inversion'
    """
    # Base waves
    p_wave = np.exp(-((t - (beat_start + 0.1))**2) / (2 * (0.025**2))) * 0.1
    qrs = np.exp(-((t - (beat_start + 0.2))**2) / (2 * (0.01**2))) * 1.0
    t_wave = np.exp(-((t - (beat_start + 0.35))**2) / (2 * (0.05**2))) * 0.3

    st_segment_mask = (t >= (beat_start + 0.22)) & (t <= (beat_start + 0.34))
    st_segment = np.zeros_like(t)

    if condition == 'normal':
        beat = p_wave - qrs + t_wave
    elif condition == 'arrhythmia':
        jitter = np.random.uniform(-0.05, 0.05)
        p_wave = np.exp(-((t - (beat_start + 0.1 + jitter))**2) / (2 * (0.03**2))) * (0.1 + 0.05*np.random.rand())
        qrs = np.exp(-((t - (beat_start + 0.2 + jitter))**2) / (2 * (0.015**2))) * (1.0 + 0.5*np.random.rand())
        t_wave = np.exp(-((t - (beat_start + 0.35 + jitter))**2) / (2 * (0.07**2))) * (0.3 + 0.2*np.random.rand())
        beat = p_wave - qrs + t_wave
    elif condition == 'st_depression':
        st_segment[st_segment_mask] = -0.15
        beat = p_wave - qrs + t_wave + st_segment
    elif condition == 'st_elevation':
        st_segment[st_segment_mask] = 0.15
        beat = p_wave - qrs + t_wave + st_segment
    elif condition == 't_wave_inversion':
        beat = p_wave - qrs - t_wave  # invert T wave by negation
    else:
        beat = p_wave - qrs + t_wave  # fallback normal

    return beat

def generate_synthetic_ecg_sample(duration=10, fs=300, heart_rate=60, condition='normal', noise_level=0.02):
    t = np.linspace(0, duration, int(duration * fs))
    beats_per_sec = heart_rate / 60
    beat_period = 1 / beats_per_sec
    ecg = np.zeros_like(t)

    beat_times = np.arange(0, duration, beat_period)

    if condition == 'arrhythmia':
        beat_times = beat_times + np.random.uniform(-0.1, 0.1, size=beat_times.shape)
        beat_times = np.clip(beat_times, 0, duration)
        beat_times = np.sort(beat_times)

    for bt in beat_times:
        ecg += generate_ecg_beat(t, bt, fs, condition=condition)

    # Add baseline wander and noise
    baseline = 0.05 * np.sin(2 * np.pi * 0.5 * t)
    noise = np.random.normal(0, noise_level, len(t))
    ecg += baseline + noise

    return t, ecg

def generate_dataset(n_samples_per_class=20, duration=10, fs=300):
    conditions = ['normal', 'arrhythmia', 'st_depression', 'st_elevation', 't_wave_inversion']
    label_map = {cond: idx for idx, cond in enumerate(conditions)}
    X, y = [], []
    for cond in conditions:
        for _ in range(n_samples_per_class):
            hr = np.random.randint(60, 101)
            _, ecg = generate_synthetic_ecg_sample(duration=duration, fs=fs, heart_rate=hr, condition=cond)
            X.append(ecg)
            y.append(label_map[cond])
    return np.array(X), np.array(y), label_map

# Example usage:
X, y, label_map = generate_dataset(n_samples_per_class=5, duration=5, fs=300)

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 8))
for i in range(10):
    plt.subplot(5, 2, i+1)
    plt.plot(X[i])
    label_name = list(label_map.keys())[list(label_map.values()).index(y[i])]
    plt.title(f'Sample {i+1} - {label_name}')
    plt.grid(True)
plt.tight_layout()
plt.show()
```
